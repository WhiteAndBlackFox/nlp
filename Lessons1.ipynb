### Созданные функции


```python
# Замена по регулярному выражению на текст
def replace_by_regexp(txt, regexp, new_val):
  return re.sub(regexp, new_val, str(txt))
```

```python
# Изменим регистр твитов на нижний
def lower_text(txt):
  return str(txt).lower()
```

```python
# Замена слов по регулярному вырадению из словаря
def replace_by_dict(txt, selected_dict, regexp = "\w*"):
  for abbreviations in re.findall(regexp, str(txt)):
    if abbreviations in selected_dict:
      txt = str(txt).replace(abbreviations, selected_dict[abbreviations])
  return txt
```

# Запуск данных

```python
def print_data(comment, data):
  print('-' * 40)
  print(comment)
  print('-' * 40)
  print(data)
  print('-' * 40)
```

```python
def run_lessons(train_df, test_df, count_print=5):
  # 1. remove user
  # train
  train_df['clean_tweet'] = train_df['tweet'].apply(lambda txt: replace_by_regexp(txt, "@[\w]*", ""))
  print_data('Delete user in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet'] = test_df['tweet'].apply(lambda txt: replace_by_regexp(txt, "@[\w]*", ""))
  print_data('Delete user in test data:', test_df.head(count_print))

  # 2. Change the case of tweets to lowercase
  # There are several ways to solve this problem 
  # First method in train data
  train_df['clean_tweet'] = train_df['clean_tweet'].apply(lower_text)
  print_data('Lower case in train data:', train_df.head(count_print))

  # Second method in test data
  test_df['clean_tweet'] = test_df['clean_tweet'].apply(lambda txt: str(txt).lower())
  print_data('Lower case in train data:', test_df.head(count_print))

  # 3. Replace abbreviations with apostrophes
  # train
  train_df['clean_tweet'] = train_df['clean_tweet'].apply(lambda txt: replace_by_dict(txt, apostrophe_dict, "\w{1,}\'\w{1,}"))
  print_data('Replace apostrophes in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet'] = test_df['clean_tweet'].apply(lambda txt: replace_by_dict(txt, apostrophe_dict, "\w{1,}\'\w{1,}"))
  print_data('Replace apostrophes in test data:', test_df.head(count_print))

  # 4. Replace abbreviations with their full forms
  # train
  train_df['clean_tweet'] = train_df['clean_tweet'].apply(lambda txt: replace_by_dict(txt, short_word_dict))
  print_data('Replace abbreviations with their full forms in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet'] = test_df['clean_tweet'].apply(lambda txt: replace_by_dict(txt, short_word_dict))
  print_data('Replace abbreviations with their full forms in test data:', test_df.head(count_print))

  # 5. Replace emoticons
  # train
  train_df['clean_tweet'] = train_df['clean_tweet'].apply(lambda txt: replace_by_dict(txt, emoticon_dict, "[:()<>=\[\]\{\}\-\‑83^|co]{2,}"))
  print_data('Replace emoticons in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet'] = test_df['clean_tweet'].apply(lambda txt: replace_by_dict(txt, emoticon_dict, "[:()<>=\[\]\{\}\-\‑83^|co]{2,}"))
  print_data('Replace emoticons in test data:', test_df.head(count_print))

  # 6. Replace punctuation with spaces
  # train
  train_df['clean_tweet'] = train_df['clean_tweet'].apply(lambda txt: replace_by_regexp(txt, "[^\w\s]", " "))
  print_data('Replace punctuation with spaces in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet'] = test_df['clean_tweet'].apply(lambda txt: replace_by_regexp(txt, "[^\w\s]", " "))
  print_data('Replace punctuation with spaces in test data:', test_df.head(count_print))

  # 7. Replace special characters with spaces
  # train
  train_df['clean_tweet'] = train_df['clean_tweet'].apply(lambda txt: replace_by_regexp(txt, "[^a-zA-Z0-9]", " "))
  print_data('Replace special characters with spaces in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet'] = test_df['clean_tweet'].apply(lambda txt: replace_by_regexp(txt, "[^a-zA-Z0-9]", " "))
  print_data('Replace special characters with spaces in test data:', test_df.head(count_print))

  # 8. Replace numbers with spaces
  # train
  train_df['clean_tweet'] = train_df['clean_tweet'].apply(lambda txt: replace_by_regexp(txt, "[^a-zA-Z]", " "))
  print_data('Replace numbers with spaces in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet'] = test_df['clean_tweet'].apply(lambda txt: replace_by_regexp(txt, "[^a-zA-Z]", " "))
  print_data('Replace numbers with spaces in test data:', test_df.head(count_print))

  # 9. Remove words with a length of 1 character from the text
  # Several solutions:
  # 1.built functions in python
  # 2.regular expressions
  # Result:
  # The first option is faster with a result of 0.09800386428833008. 
  # The second option is slower with a result of 0.1324615478515625.
  # train
  train_df['clean_tweet'] = train_df['clean_tweet'].apply(lambda txt: ' '.join([w for w in str(txt).split() if len(w)>1]))
  print_data('Remove words with a length of 1 character from the text in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet'] = test_df['clean_tweet'].apply(lambda txt: ' '.join(re.findall("\w{2,}", str(txt))))
  print_data('Remove words with a length of 1 character from the text in test data:', test_df.head(count_print))

  # 10. Divide tweets into tokens
  # train
  train_df['clean_tweet_token'] = train_df['clean_tweet'].apply(lambda txt: word_tokenize(str(txt)))
  print_data('Divide tweets into tokens in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet_token'] = test_df['clean_tweet'].apply(lambda txt: word_tokenize(str(txt)))
  print_data('Divide tweets into tokens in test data:', test_df.head(count_print))

  # 11. Remove stop words from tokens
  # train
  train_df['clean_tweet_token_stop'] = train_df['clean_tweet_token'].apply(lambda txt : [w for w in txt if not w in stop_words])
  print_data('Remove stop words from tokens in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet_token_stop'] = test_df['clean_tweet_token'].apply(lambda txt : [w for w in txt if not w in stop_words])
  print_data('Remove stop words from tokens in test data:', test_df.head(count_print))

  # 12. Apply stemming to tokens
  stemmer = PorterStemmer()
  # train
  train_df['clean_tweet_token_stemmed'] = train_df['clean_tweet_token_stop'].apply(lambda txt : [stemmer.stem(w) for w in str(txt)])
  print_data('Apply stemming to tokens in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet_token_stemmed'] = test_df['clean_tweet_token_stop'].apply(lambda txt : [stemmer.stem(w) for w in str(txt)])
  print_data('Apply stemming to tokens in test data:', test_df.head(count_print))

  # 13. Apply lemmatization to tokens
  lemm = WordNetLemmatizer()
  # train
  train_df['clean_tweet_token_lemmed'] = train_df['clean_tweet_token_stop'].apply(lambda txt : [ lemm.lemmatize(w, wordnet.VERB) for w in str(txt) ])
  print_data('Apply lemmatization to tokens in train data:', train_df.head(count_print))

  # test
  test_df['clean_tweet_token_lemmed'] = test_df['clean_tweet_token_stop'].apply(lambda txt : [ lemm.lemmatize(w, wordnet.VERB) for w in str(txt) ])
  print_data('Apply lemmatization to tokens in test data:', test_df.head(count_print))

  # 14. Save 2 pickle
  # train
  train_df.to_pickle('gdrive/My Drive/Colab Notebooks/nlp/Lessons 1 - Prepare text/train_df.pkl')

  # test
  test_df.to_pickle('gdrive/My Drive/Colab Notebooks/nlp/Lessons 1 - Prepare text/test_df.pkl')
```
